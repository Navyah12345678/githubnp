```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x=np.arange(0,math.pi*2,0.05)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.grid(True)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine wave")
plt.xticks([0,1,2,3,4,5,6],["0pi","1pi","2pi","3pi","4pi","5pi","6pi"])
plt.show()
```

```python
#pylab module
from numpy import *
from pylab import *
x=np.linspace(-3,3,30)
plot(x,sin(x),'r.',label="sin")
plot(x,cos(x),'g.',label="cos")
legend()
show()
```

```python
fig,ax=plt.subplots()
a=np.arange(0,math.pi*2)
ax.plot(a,-sin(a))
ax.grid(True)
plt.show()
```
